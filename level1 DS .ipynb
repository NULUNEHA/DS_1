```python
import plotly.express as px
fig=px.scatter_mapbox(dataset,lat="Latitude",lon="Longitude",hover_name="Restaurant Name",hover_data=["City","Cuisines"])
fig.update_layout(mapbox_style ="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
